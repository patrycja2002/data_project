# What countries receive the most medals nowadays? Does it change much across the timespan?

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('data_project/athlete_events.csv')
```

```python
df.head()
```

```python
new_df = df[["NOC","Year", "Medal"]]
```

```python
new_df
```

```python
cleaning_new_df = new_df.dropna(axis="index", how="any", subset=["Medal", "NOC", "Year"])
```

```python
cleaning_new_df
```

```python
new_data_frame = cleaning_new_df.drop(columns=["Medal"])
new_data_frame
```

```python
year_df = new_data_frame.value_counts("Year")
dictionary = year_df.to_dict()
year = []
year_df = [year for year in dictionary.keys()]
```

```python
year_df.sort()
year_df
def year_(year):
    for year_df in year:
        filt = (new_data_frame["Year"] == year_df)
        data_frame = new_data_frame.loc[filt]["NOC"].value_counts().head()
        print(year_df)
        print(data_frame)
print(year_(year_df))
    
    
    
```
